```python
from ultralytics import YOLO
import cv2

# 학습된 모델 불러오기
model = YOLO("best.pt")

# 웹캠 열기
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    if not ret:
        print("웹캠을 읽을 수 없습니다.")
        break

    # YOLO 예측
    results = model.predict(
        frame,
        imgsz=640,
        conf=0.5,
        verbose=False
    )

    # 결과 이미지 생성
    annotated_frame = results[0].plot()

    # 화면 출력
    cv2.imshow("YOLO Webcam Result", annotated_frame)

    # q 누르면 종료
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
