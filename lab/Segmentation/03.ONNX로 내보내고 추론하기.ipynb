```python
import cv2
from ultralytics import YOLO

model = YOLO("yolo26n-seg.onnx")  # pt 대신 onnx 로드

cap = cv2.VideoCapture(0)

if cap.isOpened():
    print("실시간 세그멘테이션 시작")
    print("종료하려면 'q'를 누르세요.")

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        results = model(frame, conf=0.25, iou=0.45, verbose=False)
        annotated_frame = results[0].plot()

        cv2.imshow("YOLO26 Segmentation Realtime", annotated_frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

cap.release()
cv2.destroyAllWindows()
```
