## 파라미터 탐색한 결과에 freeze=10 추가 후
- head만 훈련

```python
from ultralytics import YOLO
import yaml
 
# YOLOv8 모델 로드
model = YOLO("yolov8n.pt")
 
# 하이퍼파라미터 yaml 파일 불러오기
with open("runs/detect/tune/best_hyperparameters.yaml", "r", encoding="utf-8") as f:
    hyp_dict = yaml.safe_load(f)
 
# 모델 훈련
model.train(data="./datasets/data.yaml", epochs=250, imgsz=640, **hyp_dict)
```
