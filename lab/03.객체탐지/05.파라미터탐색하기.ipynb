## 하이퍼 파라미터 튜닝 - fine tunning의 일부
- 250 에폭 25번 돌려서 최적의 파라미터 찾기  
- tune/weights/best_hyperparametgers.yaml로 생성
- 베스트 모델은 best.pt

```python
from ultralytics import YOLO
 
# YOLOv8 모델 로드
model = YOLO("yolov8n.pt")
 
# 하이퍼파라미터 탐색 수행
model.tune(
    data="./datasets/data.yaml",
    epochs=250,
    iterations=25, # 탐색 횟수 (예: 25번의 실험)
    val=True
)
```
